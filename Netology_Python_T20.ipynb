```python
import random
import time
import http.client
import json
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

from scipy.cluster.vq import whiten, kmeans2

MY_API = '2bccc5e3c4ee482ab03922ac8c5f358a'
CONNECTION = http.client.HTTPConnection('api.football-data.org')
HEADERS = { 'X-Auth-Token': MY_API, 'X-Response-Control': 'minified' }
connect_to_api = (CONNECTION, HEADERS)
```

```python
def get_competitions_api_football(params, first_year, last_year): 
    competitions = pd.DataFrame({})
    
    for year in range(first_year, last_year + 1):
        response = get_any_api_football(params.format(year))
        competitions_of_year = pd.DataFrame(response)
        competitions = pd.concat([competitions, competitions_of_year])
    
    return competitions
```

```python
def get_any_api_football(params, result_key=None):
    response = {}
    connection, headers = connect_to_api
    connection.request('GET', params, None, headers)
    try:
        response = json.loads(connection.getresponse().read().decode())
        result = response if result_key is None else response[result_key] 
        return result
    except:
        return None
```

```python
competitions = get_competitions_api_football('/v1/competitions/?season={}', 2015, 2015)
```

```python
def get_data_api_football(data_uri, data_key):
    get_data = None
    
    count_errors = 0
    max_errors = 1
    
    while (get_data is None) and (count_errors <= max_errors): 
        get_data = get_any_api_football(data_uri, data_key)
        if  get_data is None:
            count_errors += 1
            time.sleep(60)
  
    return get_data


def get_team_looses(team_name, fixtures):    
    team_looses = 0
    for fixture in fixtures:
        home_goals = fixture['result']['goalsHomeTeam']
        away_goals = fixture['result']['goalsAwayTeam']

        goals_team =  home_goals if team_name == fixture['homeTeamName'] else away_goals  
        goals_enemy = away_goals if team_name == fixture['homeTeamName'] else home_goals
                    
        if goals_team == goals_enemy:
            if fixture['result'].get('penaltyShootout'):
                home_penalty = fixture['result']['penaltyShootout']['goalsHomeTeam']
                away_penalty = fixture['result']['penaltyShootout']['goalsAwayTeam']
                goals_team =  home_penalty if team_name == fixture['homeTeamName'] else away_penalty  
                goals_enemy = away_penalty if team_name == fixture['homeTeamName'] else home_penalty
                
        if isinstance(goals_team, int):
            if goals_team < goals_enemy:
                team_looses += 1  
    return team_looses


def get_teams_api_football(competitions): 
    teams_for_pd = []
    min_price = 100000
    max_price = 100000000
    
    teams_key = 'teams'   
    fixture_key = 'fixtures'
    
    id_competitions = list(competitions['id'])
    
    for id in id_competitions:
        teams_uri = '/v1/competitions/{}/teams'.format(id)
        teams = get_data_api_football(teams_uri, teams_key)
    
        if not (teams is None):
            for team in teams:
                fixture_uri = '/v1/teams/{}/fixtures'.format(team['id'])
                fixtures = get_any_api_football(fixture_uri, fixture_key)
                
                if not (fixtures is None):
                    team_looses = get_team_looses(team['name'], fixtures)    
                    random_price = random.randint(min_price, max_price)   
                    team_price = random_price if team['squadMarketValue'] == None else team['squadMarketValue']  
                    teams_for_pd.append({'id': team['id'], 'name': team['name'], 'price': team_price, 'looses': team_looses})
            
    teams_pd = pd.DataFrame(teams_for_pd)
    print('Dataset is ready!')
    return teams_pd
```

```python
teams_pd = get_teams_api_football(competitions)
```
